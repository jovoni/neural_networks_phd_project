```python
import matplotlib.pyplot as plt
import pandas as pd
import sys
sys.path.insert(0, "src/")
```

```python
# Params
K = 3
model_name = "twoLNN"
batch_size = 128
task = "parity"
```

```python
path = f'results/{task}_{model_name}_{K}_{batch_size}.csv'
df = pd.read_csv(path, sep=",")
```

```python
x = df.epoch
y_test = df.test_acc
y_train = df.train_acc

plt.plot(x, y_train, label="Train")
plt.plot(x, y_test, label="Test")
plt.legend()
plt.ylim(0.5,1.01)
plt.show()
```
