```python
import matplotlib.pyplot as plt
import torch
import pandas as pd
import sys
sys.path.insert(0, "src/")
```

```python
# Params
K = 1
model_name = "twoLNN"
lr = 0.05
nepochs = 20
batch_size = 128
task_name = "parity"
```

```python
path = f'results/{task_name}_{model_name}_{K}_{lr}_{batch_size}_{nepochs}.csv'
df = pd.read_csv(path, sep=",")
```

```python
x = df.epoch
y_test = df.test_acc
y_train = df.train_acc

plt.plot(x, y_train, label="Train")
plt.plot(x, y_test, label="Test")
plt.legend()
plt.ylim(0.4,1.01)
plt.xticks([0,5,10,15])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```
